```python
from __future__ import division
from math import log
import math  
from random import shuffle 

def entropy(pi):
    total = 0
    for p in pi:
        p = p / sum(pi)
        if p != 0:
            total += p * log(p, 2)
        else:
            total += 0
    total *= -1
    return total


def gain(d, a):
    total = 0
    for v in a:
        total += sum(v) / sum(d) * entropy(v)

    gain = entropy(d) - total
    return gain

def symmetric_uncertainity(X,Y):
    su = 2*gain(X,Y)/(entropy(X)+entropy(Y))
    return su
```
